## üì¶ Setup & Dependencies

First, let's import all necessary libraries and set up our environment:

````python
import sys
import os
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
import json

# Configure plotting
plt.style.use('default')
sns.set_palette("husl")
%matplotlib inline

# Add FlowerPower src to path
sys.path.insert(0, str(Path.cwd().parents[2] / "src"))

print("‚úÖ Dependencies loaded successfully!")
print(f"üìÇ Current directory: {Path.cwd()}")
````

## üîß Initialize FlowerPower Pipeline

Let's set up the FlowerPower pipeline manager to run our ETL pipeline:

````python
from flowerpower.pipeline.manager import PipelineManager

# Initialize pipeline manager
pipeline_manager = PipelineManager(
    base_dir=".",
    fs=None
)

print("‚úÖ FlowerPower pipeline manager initialized!")
print(f"üìã Available pipelines: {pipeline_manager.list_pipelines()}")
````

## üìä Explore Sample Data

Before running the pipeline, let's examine our sample sales data:

````python
# Load the raw sales data
sales_data = pd.read_csv("data/sales_data.csv")

print(f"üìä Dataset shape: {sales_data.shape}")
print(f"üìÖ Date range: {sales_data['date'].min()} to {sales_data['date'].max()}")
print(f"üõçÔ∏è Unique products: {sales_data['product'].nunique()}")
print(f"üë• Unique customers: {sales_data['customer'].nunique()}")

# Display first few rows
print("\nüîç Sample data:")
sales_data.head()
````

````python
# Data quality overview
print("üìã Data Quality Summary:")
print(f"Missing values: {sales_data.isnull().sum().sum()}")
print(f"Duplicate rows: {sales_data.duplicated().sum()}")

# Basic statistics
print("\nüìà Basic Statistics:")
print(sales_data.describe())
````

## üéØ Run ETL Pipeline - Step by Step

Now let's run our ETL pipeline and examine the results at each step:

````python
# Run the complete ETL pipeline
print("üîÑ Running ETL pipeline...")

result = pipeline_manager.run(
    "sales_etl",
    inputs={},  # Using defaults from config
    final_vars=["clean_data", "sales_summary", "validation_report", "processed_file_path"]
)

print("‚úÖ Pipeline execution completed!")
print(f"üìä Results available: {list(result.keys())}")
````

### üîç Validation Report Analysis

````python
# Examine validation report
validation_report = result['validation_report']

print("üìã Data Validation Report:")
print(f"‚úÖ Data is valid: {validation_report['is_valid']}")
print(f"üìä Total records: {validation_report['total_records']}")
print(f"üí∞ Price violations: {validation_report['price_violations']}")

if validation_report['missing_columns']:
    print(f"‚ùå Missing columns: {validation_report['missing_columns']}")
    
if validation_report['data_quality_issues']:
    print(f"‚ö†Ô∏è Quality issues:")
    for issue in validation_report['data_quality_issues']:
        print(f"  - {issue}")

# Missing values breakdown
if validation_report['missing_values']:
    print(f"\nüï≥Ô∏è Missing values by column:")
    for col, count in validation_report['missing_values'].items():
        print(f"  - {col}: {count} missing")
````

### üßπ Clean Data Analysis

````python
# Examine cleaned data
clean_data = result['clean_data']

print(f"üìä Clean data shape: {clean_data.shape}")
print(f"üßπ Data types after cleaning:")
print(clean_data.dtypes)

# Check for any remaining data quality issues
print(f"\n‚úÖ Missing values after cleaning: {clean_data.isnull().sum().sum()}")
print(f"üìà Total revenue: ${clean_data['total_sales'].sum():,.2f}")

# Display sample of cleaned data
print("\nüîç Sample of cleaned data:")
clean_data.head()
````

### üìà Sales Summary Analysis

````python
# Examine sales summary
sales_summary = result['sales_summary']

print(f"üìä Sales summary shape: {sales_summary.shape}")
print(f"üõçÔ∏è Product groups: {sales_summary['product'].nunique()}")
print(f"üë• Customer groups: {sales_summary['customer'].nunique()}")

# Top performers
print("\nüèÜ Top 10 Product-Customer combinations by revenue:")
top_combinations = sales_summary.nlargest(10, 'total_sales')
for _, row in top_combinations.iterrows():
    print(f"  {row['product']} + {row['customer']}: ${row['total_sales']:,.2f} ({row['total_quantity']} items)")

# Display full summary
print("\nüìä Sales Summary Data:")
sales_summary.head(10)
````

## üìä Data Visualization & Analysis

Let's create some visualizations to better understand our data:

````python
# Set up the plotting area
fig, axes = plt.subplots(2, 2, figsize=(15, 12))
fig.suptitle('üìä Sales Data Analysis Dashboard', fontsize=16, fontweight='bold')

# 1. Revenue by Product
product_revenue = sales_summary.groupby('product')['total_sales'].sum().sort_values(ascending=False)
axes[0, 0].bar(range(len(product_revenue)), product_revenue.values)
axes[0, 0].set_title('üí∞ Revenue by Product')
axes[0, 0].set_xlabel('Products')
axes[0, 0].set_ylabel('Total Revenue ($)')
axes[0, 0].tick_params(axis='x', rotation=45)

# 2. Sales Distribution
axes[0, 1].hist(clean_data['total_sales'], bins=30, alpha=0.7, color='skyblue')
axes[0, 1].set_title('üíµ Sale Amount Distribution')
axes[0, 1].set_xlabel('Sale Amount ($)')
axes[0, 1].set_ylabel('Frequency')

# 3. Customer Spending
customer_spending = sales_summary.groupby('customer')['total_sales'].sum().sort_values(ascending=False)
axes[1, 0].bar(range(len(customer_spending)), customer_spending.values)
axes[1, 0].set_title('üë• Customer Total Spending')
axes[1, 0].set_xlabel('Customers')
axes[1, 0].set_ylabel('Total Spending ($)')

# 4. Quantity vs Revenue Scatter
axes[1, 1].scatter(sales_summary['total_quantity'], sales_summary['total_sales'], alpha=0.6)
axes[1, 1].set_title('üì¶ Quantity vs Revenue')
axes[1, 1].set_xlabel('Total Quantity')
axes[1, 1].set_ylabel('Total Sales ($)')

plt.tight_layout()
plt.show()
````

````python
# Time series analysis (if date column is available)
if 'date' in clean_data.columns:
    # Convert date and create time-based analysis
    clean_data['date'] = pd.to_datetime(clean_data['date'])
    daily_sales = clean_data.groupby('date')['total_sales'].sum()
    
    plt.figure(figsize=(12, 6))
    plt.plot(daily_sales.index, daily_sales.values, marker='o', linewidth=2)
    plt.title('üìÖ Daily Sales Trend', fontsize=14, fontweight='bold')
    plt.xlabel('Date')
    plt.ylabel('Daily Revenue ($)')
    plt.xticks(rotation=45)
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.show()
    
    print(f"üìà Average daily revenue: ${daily_sales.mean():,.2f}")
    print(f"üîù Highest revenue day: {daily_sales.idxmax()} (${daily_sales.max():,.2f})")
    print(f"üìâ Lowest revenue day: {daily_sales.idxmin()} (${daily_sales.min():,.2f})")
````

## ‚öôÔ∏è Pipeline Configuration Experiments

Let's experiment with different pipeline configurations to see how they affect the results:

### üß™ Experiment 1: Different Validation Thresholds

````python
# Test with stricter validation thresholds
print("üß™ Experiment 1: Stricter validation thresholds")

strict_result = pipeline_manager.run(
    "sales_etl",
    inputs={
        "min_price": 5.0,   # Higher minimum price
        "max_price": 500.0  # Lower maximum price
    },
    final_vars=["validation_report"]
)

strict_validation = strict_result['validation_report']
print(f"‚úÖ Strict validation result: {strict_validation['is_valid']}")
print(f"üí∞ Price violations: {strict_validation['price_violations']}")
print(f"‚ö†Ô∏è Quality issues: {len(strict_validation['data_quality_issues'])}")

# Compare with original
print(f"\nüìä Comparison:")
print(f"Original price violations: {validation_report['price_violations']}")
print(f"Strict price violations: {strict_validation['price_violations']}")
````

### üß™ Experiment 2: Different Aggregation Rules

````python
# Test with different aggregation grouping
print("üß™ Experiment 2: Product-only aggregation")

product_only_result = pipeline_manager.run(
    "sales_etl",
    inputs={
        "group_by": ["product"],  # Group by product only
        "metrics": ["total_sales", "avg_price", "total_quantity"]
    },
    final_vars=["sales_summary"]
)

product_summary = product_only_result['sales_summary']
print(f"üìä Product-only summary shape: {product_summary.shape}")
print(f"üõçÔ∏è Products analyzed: {product_summary['product'].nunique()}")

print("\nüèÜ Top 5 products by revenue:")
top_products = product_summary.nlargest(5, 'total_sales')
for _, row in top_products.iterrows():
    print(f"  {row['product']}: ${row['total_sales']:,.2f} (avg: ${row['avg_price']:.2f})")

# Visualize product comparison
plt.figure(figsize=(10, 6))
plt.bar(top_products['product'], top_products['total_sales'])
plt.title('üèÜ Top 5 Products by Revenue')
plt.xlabel('Product')
plt.ylabel('Total Revenue ($)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
````

### üß™ Experiment 3: Custom Date Range Processing

````python
# Analyze recent vs older sales patterns
print("üß™ Experiment 3: Date-based analysis")

# Create separate analyses for different time periods
if 'date' in clean_data.columns:
    clean_data['date'] = pd.to_datetime(clean_data['date'])
    
    # Split data into two periods
    mid_date = clean_data['date'].quantile(0.5)
    
    early_period = clean_data[clean_data['date'] <= mid_date]
    late_period = clean_data[clean_data['date'] > mid_date]
    
    print(f"üìÖ Early period: {early_period['date'].min()} to {early_period['date'].max()}")
    print(f"üìÖ Late period: {late_period['date'].min()} to {late_period['date'].max()}")
    
    # Compare metrics
    early_revenue = early_period['total_sales'].sum()
    late_revenue = late_period['total_sales'].sum()
    
    early_avg = early_period['total_sales'].mean()
    late_avg = late_period['total_sales'].mean()
    
    print(f"\nüí∞ Revenue comparison:")
    print(f"  Early period: ${early_revenue:,.2f} (avg: ${early_avg:.2f})")
    print(f"  Late period: ${late_revenue:,.2f} (avg: ${late_avg:.2f})")
    print(f"  Growth: {((late_revenue - early_revenue) / early_revenue * 100):+.1f}%")
    
    # Visualize comparison
    plt.figure(figsize=(10, 6))
    periods = ['Early Period', 'Late Period']
    revenues = [early_revenue, late_revenue]
    
    plt.bar(periods, revenues, color=['skyblue', 'lightcoral'])
    plt.title('üìä Revenue Comparison: Early vs Late Period')
    plt.ylabel('Total Revenue ($)')
    
    # Add value labels on bars
    for i, v in enumerate(revenues):
        plt.text(i, v + max(revenues) * 0.01, f'${v:,.0f}', ha='center', va='bottom')
    
    plt.tight_layout()
    plt.show()
else:
    print("‚ö†Ô∏è Date column not available for time-based analysis")
````

## üîß Advanced Pipeline Customization

Let's explore how to customize the pipeline for specific business needs:

````python
# Create a custom analysis with multiple configurations
print("üîß Advanced Pipeline Customization")

# Test different configurations
configurations = [
    {
        "name": "Standard Analysis",
        "config": {
            "group_by": ["product", "customer"],
            "min_price": 0,
            "max_price": 10000
        }
    },
    {
        "name": "High-Value Focus",
        "config": {
            "group_by": ["product"],
            "min_price": 50,
            "max_price": 10000
        }
    },
    {
        "name": "Customer Focus",
        "config": {
            "group_by": ["customer"],
            "min_price": 0,
            "max_price": 10000
        }
    }
]

results_comparison = []

for config in configurations:
    print(f"\nüîÑ Running {config['name']}...")
    
    result = pipeline_manager.run(
        "sales_etl",
        inputs=config['config'],
        final_vars=["sales_summary", "validation_report"]
    )
    
    summary = result['sales_summary']
    validation = result['validation_report']
    
    analysis = {
        "name": config['name'],
        "groups": len(summary),
        "total_revenue": summary['total_sales'].sum(),
        "avg_revenue_per_group": summary['total_sales'].mean(),
        "price_violations": validation['price_violations'],
        "is_valid": validation['is_valid']
    }
    
    results_comparison.append(analysis)
    
    print(f"  ‚úÖ Groups created: {analysis['groups']}")
    print(f"  üí∞ Total revenue: ${analysis['total_revenue']:,.2f}")
    print(f"  üìä Avg per group: ${analysis['avg_revenue_per_group']:,.2f}")

# Create comparison DataFrame
comparison_df = pd.DataFrame(results_comparison)
print("\nüìä Configuration Comparison:")
comparison_df
````

````python
# Visualize configuration comparison
fig, axes = plt.subplots(1, 2, figsize=(15, 6))

# Total revenue comparison
axes[0].bar(comparison_df['name'], comparison_df['total_revenue'])
axes[0].set_title('üí∞ Total Revenue by Configuration')
axes[0].set_ylabel('Total Revenue ($)')
axes[0].tick_params(axis='x', rotation=45)

# Groups created comparison
axes[1].bar(comparison_df['name'], comparison_df['groups'])
axes[1].set_title('üìä Number of Groups Created')
axes[1].set_ylabel('Number of Groups')
axes[1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
````

Let's save our analysis results for future use:
